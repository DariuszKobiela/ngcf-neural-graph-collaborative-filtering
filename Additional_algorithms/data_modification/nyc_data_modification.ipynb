```python
import pandas as pd
```

```python

with open('test.txt') as f:
    lines = [line.rstrip() for line in f]

df = pd.DataFrame(columns=['User', "Item", "Rating"])

lines_num = len(lines)
it = 1

for line in lines:
    print(f"Line number {it} out of {lines_num} of length {len(line)}")
    it+=1
    line_array = line.split()
    user_id = line_array[0]
    
    for item_id in line_array[1:]:
        d = {"User": f"user_{user_id}", "Item": f"item_{item_id}", "Rating": 1}
        df_dictionary = pd.DataFrame([d])
        df = pd.concat([df, df_dictionary], ignore_index=True)
df.to_csv("nyc_test.txt", index=False, header=None, sep="\t")
df
```

```python
test = pd.read_csv("nyc_test.txt", sep="\t", header=None,)
train = pd.read_csv("nyc_train.txt", sep="\t", header=None)

df = pd.concat([test, train], axis=0)
df
```

```python
df.rename(columns={0: 'User',1: 'Item',2: 'Rating'},
            inplace=True)
df
```

```python
user_array = df["User"].unique()
item_array = df["Item"].unique()
```

```python
field_df = pd.DataFrame(columns=["ID", "Field"])

it = 0
user_num = len(user_array)

for user in user_array:
    print(f"User {it} out of {user_num}")
    it+=1
    d = {"ID": user, "Field": "u"}
    df_dictionary = pd.DataFrame([d])
    field_df = pd.concat([field_df, df_dictionary], ignore_index=True)

it = 0
itemr_num = len(item_array)

for item in item_array:
    print(f"Item {it} out of {itemr_num}")
    d = {"ID": item, "Field": "i"}
    df_dictionary = pd.DataFrame([d])
    field_df = pd.concat([field_df, df_dictionary], ignore_index=True)
    
field_df.to_csv("nyc_field.txt", sep="\t", index=False, header=None)
```
